```python
import pandas as pd
import csv, sqlite3

df1 = pd.read_csv("Covid_Vaccine_data_state_formatted.csv")
df2 = pd.read_csv("Covid_Vaccine_data_state_YE2021_formatted.csv")
df3 = pd.read_csv("Covid_Vaccine_data_US_formatted.csv")
df4 = pd.read_csv("COVID-19_Vaccinations_in_the_United_States_Jurisdiction formatted.csv")

conn = sqlite3.connect("covid_dashboard.sqlite") 

df1.to_sql("VaccineState", conn, if_exists='append', index=False)
df2.to_sql("VaccineStateYearEnd", conn, if_exists='append', index=False)
df3.to_sql("VaccineUS", conn, if_exists='append', index=False)
df4.to_sql("VaccineRaw", conn, if_exists='append', index=False)
```

```python
cur = conn.cursor()

# Return all results of query
cur.execute('SELECT * FROM VaccineState')
cur.fetchall()

# Be sure to close the connection
# conn.close()
```
